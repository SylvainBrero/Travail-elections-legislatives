```python
## On peut généraliser cette méthode relativement facilement à l'échelle du département / région / pays, sans se soucier du bureau de vote 
## et des résultats des candidats
## Pour notre analyse méta, on va commencer par procéder ainsi

## Remarque : On sera obligé de garder la colonne "Code du b.vote" car elle permettra, pour la ville de Lyon uniquement, de récupérer le code INSEE de l'arrondissement correspondant
## On la renommera "Arrondissement"
```

```python
df_rhone.head()
```

```python
df_rhone_general =  df_rhone.iloc[:, 5:21]
df_rhone_general.head()
```

```python
## Le code du bureau de vote est composé de : 
## 2 chiffres signifiant l'arrondissement (pour les grandes villes)
## 2 chiffres permettant la dénomination à l'intérieur de l'arrondissement

## Exemple : Code du premier bureau de la ville d'Alix : 0001
## Exemple : Code du premier bureau du 7eme Arrondissement de Lyon : 0701

## On ne garde que les 2 premiers chiffres de ce code
df_rhone_general['Code du b.vote'] = df_rhone_general['Code du b.vote'].apply(lambda x : x[:2])
df_rhone_general = df_rhone_general.rename(columns = {'Code du b.vote' : 'Arrondissement'})
df_rhone_general.head()
```

```python
df_rhone_grp = df_rhone_general.groupby(by=["Libellé de la commune","Arrondissement"]).mean().round(3)
df_rhone_grp.head()
```

```python
## On rajoute le code commune afin de pouvoir avoir une correspondance parfaite avec le fichier geoJSON et obtenir une carte propre :

## On reconstruit ce code (concaténation du code departement + code commune; ici brutalement 69 + XXX)
## Cas particulier : Lyon. Qui est définie comme étant de code 69123, mais dont le code INSEE est 69380, et son 1er arrondissement est 69381, le 2eme 69382 etc.

code_geojson = 0
liste_code_geojson = []

for element in df_rhone_grp.index:
    
    nom_commune = element[0]
    arrondissement = element[1]
    
    ## Récupération du code commune
    code_indf = str(df_rhone[df_rhone["Libellé de la commune"]==nom_commune]["Code de la commune"].iloc[0])
    
    
    ## Reconstitution du code geosjon
    if nom_commune != "Lyon":
        if len(code_indf) == 1:
            code_geojson = "6900"+ code_indf
        elif len(code_indf) == 2:
            code_geojson = "690"+ code_indf
        else:
            code_geojson = "69"+ code_indf
    else:
        code_geojson = "6938"+str(arrondissement[-1])
        
    liste_code_geojson.append(code_geojson)
    
df_rhone_grp["Code de la commune"] = liste_code_geojson
```
